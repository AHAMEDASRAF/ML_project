```python
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from sklearn.preprocessing import LabelEncoder

# Make a copy of df to avoid modifying the global df for this specific operation
df_processed_new = df.copy()

# Store LabelEncoders to use for new_candidate later
encoders_new = {}
categorical_cols_new = df_processed_new.select_dtypes(include=['object']).columns.tolist()

for col in categorical_cols_new:
    # Only encode if the column is not 'Recruiter_Decision' (if it's being dropped from X later)
    # and if it was present in the original categorical columns.
    if col != 'Recruiter_Decision': # Ensure we don't encode a column that will be dropped before X creation
        le = LabelEncoder()
        df_processed_new[col] = le.fit_transform(df_processed_new[col].astype(str))
        encoders_new[col] = le # Save the fitted encoder
    else:
        # For 'Recruiter_Decision', we will not encode it as it's being removed from features.
        pass

# 1. Redefine X and y to ensure 'Is_Shortlisted' is NOT in X
# Drop 'Application_Status', 'Gender', 'Resume_ID', 'Name' and 'Recruiter_Decision'
X_new = df_processed_new.drop(["Is_Shortlisted", "Application_Status", "Gender", "Resume_ID", "Name", "Recruiter_Decision"], axis=1)
y_new = df_processed_new["Is_Shortlisted"]

# Identify indices of categorical features in X_new for CatBoost
# These must be columns present in X_new, and also were originally categorical
cat_features_indices_new = [
    X_new.columns.get_loc(col) for col in ["Location", "Skills", "Degree", "University", "Education", "Certifications"]
    if col in X_new.columns
]

# 2. Re-initialize CatBoostClassifier and split data
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new, y_new, test_size=0.2, random_state=42)
model_new = CatBoostClassifier(
    iterations=500,
    learning_rate=0.1,
    depth=6,
    loss_function="Logloss", # Ensure Logloss for binary classification
    random_seed=42,
    verbose=0, # Suppress verbose output
    cat_features=cat_features_indices_new # Tell CatBoost which features are categorical
)

# 3. Re-train the model with the correct X_train_new and y_train_new
model_new.fit(X_train_new, y_train_new)

print("Model retrained successfully without 'Recruiter_Decision' as a feature.")
```
